```python
def main():
    num = 42
    num1 = 42.0
    num2 = 42 / 9
    num3 = 42 // 9  # this is a integer division
    num4 = round(42 / 9)  # roudn up the division
    num5 = round(42 / 9, 2)  # roudn up the division with 2 decimal precisions
    num6 = 42 % 9
    num7 = int(42.9)
    num8 = float(42)  # creating a object of type float and passing constructor the value 42
    
    print(type(num), num)
    print(type(num1), num1)
    print('---------------')
    print(type(num2), num2)
    print(type(num3), num3)
    print(type(num4), num4)
    print(type(num5), num5)
    print(type(num6), num6)
    print(type(num7), num7)
    print(type(num8), num8)
    
if __name__ == "__main__":
    main()
```
