exceptions are the key method of handling errors in python<br>
to catch exceptions use "try" something, then catch an exception with "except"<br>
you may raise your own exceptions with "raise"

```python
def main():
    try:
        fh = open('lines.txt')
    except IOError as e:
        print('could not open the file', e)
    else:
        for line in fh: print(line.strip())
    
if __name__ == "__main__" : main()
```
